```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
df = pd.read_csv('pytrends_100.csv', index_col = 'Datetime')

df.head()
```

```python
index = pd.date_range("2018/02/26 00:00:00", "2018/03/05 23:00:00", freq = "1H")
df["date"]=index
df
```

```python
index = pd.date_range("2018/02/26 00:00:00", "2018/03/05 23:45:00", freq = "15min")
d = {"date": index}
new_Data = pd.DataFrame(d)
new_Data = new_Data.merge(df, how = "left", left_on = "date", right_on = "date")
new_Data = new_Data.interpolate()
new_Data
```

```python
new_Data1 = new_Data.melt(id_vars=['date'])
```
